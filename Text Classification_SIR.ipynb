# Text Classification using TensorFlow

This project performs text classification using deep learning techniques. The model is trained to classify text into different categories using TensorFlow/Keras.

```python
import tensorflow as tf
from tensorflow import keras
import tensorflow_hub as hub
import tensorflow_datasets as tfds
```

```python
train_data, val_data, test_data = tfds.load(name="imdb_reviews",
                                            split=('train', 'test[:40%]', 'test[40:]'),
                                            as_supervised=True)
```

```python
for text, label in train_data.take(3):
    print(text.numpy().decode("utf-8"))
    print("Label:", label.numpy())
    print("-----")
```

```python
max_tokens = 20000
sequence_length = 250 

vectorize = tf.keras.layers.TextVectorization(
    max_tokens=max_tokens,
    output_mode="int",
    output_sequence_length=sequence_length
)

vectorize.adapt(train_data.map(lambda x, y: x))
```

```python
model = keras.Sequential([
    vectorize,
    keras.layers.Embedding(max_tokens, 128),
    keras.layers.GlobalAveragePooling1D(),
    keras.layers.Dense(16, activation="relu"),
    keras.layers.Dense(1, activation="sigmoid")
])

model.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)

model.summary()
```

```python
model = tf.keras.Sequential([
    vectorize,                              
    tf.keras.layers.Embedding(20000, 128),  
    tf.keras.layers.GlobalAveragePooling1D(),
    tf.keras.layers.Dense(64, activation="relu"),
    tf.keras.layers.Dense(1, activation="sigmoid")
])

model.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)

model.summary()
```

```python
model.fit(train_data.shuffle(1000).batch(1000), epochs=10, validation_data=val_data.batch(100), verbose=1)
```

```python
model.evaluate(test_data.batch(100))
```
